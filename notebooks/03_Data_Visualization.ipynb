```python
import matplotlib.pyplot as plt
```

```python
# Plotting my first Graph
x = [1,2,3,4,5,6,7]
y = [1,4,9,16,25,26,29]

plt.title("My first Graph")
plt.xlabel("This is X-axis")
plt.ylabel("This is Y-axis")

plt.plot(x,y)
plt.show()
```

```python
# Improvement on the Previous Graph
x = [1,2,3,4,5,6,7]
y = [1,4,9,16,25,26,29]

plt.title("Improved Graph", {'fontsize':25, 'color':'#eeeee4'} )
plt.xlabel("X-axis", {'fontsize':25, 'color':'#eab676'})
plt.ylabel("Y-axis", {'fontsize':25, 'color':'#e16130'})

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [1,4,9,16,25,26]

a = [1,3,6,8]
b = [1,6,10,11]

plt.title("Multi-Line Graph", {'fontsize':25, 'color':'black'})
plt.xlabel("X-axis", {'fontsize': 25, 'color':'Green'})
plt.ylabel("Y-axis", {'fontsize': 25, 'color':'Green'})

plt.plot(x,y, color = 'red')
plt.plot(a,b, color = 'yellow')
```

```python
import pandas as pd

df = pd.read_csv('../petrol_prices.csv')
df.head()
```

Important Things to Remember:
1. What is the purpose of the visualization??
2. What kind of information you want to communicate ?

```python
# Multi Series Line Chart or Multiple Line Plot
plt.title('Petrol Price in NA over the years', {'fontsize':15, 'color':'red'})
plt.xlabel('Year')
plt.ylabel('Petrol Prices')

plt.plot(df['Year'], df['USA'], label='USA', color='red' )
plt.plot(df['Year'], df['Canada'], label='Canada', color='green')

plt.legend()
plt.show()
```

```python
# Multi-Series Line Plot with Data Point Markers
plt.title('Petrol Price in NA over the years', {'fontsize':15, 'color':'red'})
plt.xlabel('Year')
plt.ylabel('Petrol Prices')

plt.plot(df['Year'], df['USA'], label='USA', marker='o', color='red')
plt.plot(df['Year'], df['Canada'], label='Canada', marker='v', color='blue')

plt.legend()
plt.show()
```
