```python
# Multi Series Line Chart or Multiple Line Plot
plt.title('Petrol Price in NA over the years', {'fontsize':15, 'color':'red'})
plt.xlabel('Year')
plt.ylabel('Petrol Prices')

plt.plot(df['Year'], df['USA'], label='USA', color='red' )
plt.plot(df['Year'], df['Canada'], label='Canada', color='green')

plt.legend()
plt.show()
```

```python
# Multi-Series Line Plot with Data Point Markers
plt.title('Petrol Price in NA over the years', {'fontsize':15, 'color':'red'})
plt.xlabel('Year')
plt.ylabel('Petrol Prices')

plt.plot(df['Year'], df['USA'], label='USA', marker='o', color='red')
plt.plot(df['Year'], df['Canada'], label='Canada', marker='v', color='blue')

plt.legend()
plt.show()
```

# <center>Bar Graph</center>
A bar graph is a chart that uses rectangular bars to represent and compare categorical data. The length or height of each bar is proportional to its value. Key features include:
- Uniform bar width and consistent spacing between bars.
- Orientation can be vertical (column chart) or horizontal.
- Essential elements: Title, labeled X and Y axes, data (bars), a legend (for multiple series), a data source, and hatch patterns (for differentiation without color).

```python
# First Bar Graph Visualization
x = ['HR', 'QA', 'Developer']
y = [40000, 50000, 90000]

plt.title("Starting Salaries for Different Jobs")
plt.xlabel("Job Title")
plt.ylabel("Salary")

plt.bar(x,y)
plt.show()
```

```python
year_data = df.loc[11, 'Year']
year_data
# It selects the data of 11th row in the 'Year' column 

colors = ["#FF5733", "#33FF57", "#3357FF", "#F1C40F", "#9B59B6", "#E67E22", "#1ABC9C", "#2ECC71", "#34495E"]
hatches = ['//', '\\\\', '||', '--', '++', 'xx', 'oo', 'OO', '..', '**']

plt.figure(figsize=(20,10))

bars = plt.bar(df.columns[1:], year_data, color = colors)
# This line creates a bar chart using the dataframe's column names (excluding the first) as x-axis labels, 
# the year_data values as bar heights, and applies a custom color to each bar, storing the created bar objects for further modification.
# df.columns[1:]: This provides the labels for the x-axis.
# [1:] is a slice that means "from the column at position 1 to the end." It skips the first column (index 0). 

for bar, hatch in zip(bars, hatches): # Applying hatch pattern for each bar
    bar.set_hatch(hatch)

plt.title('Data for the years')
plt.xlabel('Countries')
plt.ylabel('Prices')

plt.show()
```

# <center>Boxplot</center>
A boxplot is a standardized way of displaying the distribution of a dataset based on a five-number summary:
- Minimum
- First Quartile (Q1)
- Median (Q2)
- Third Quartile (Q3)
- Maximum

It shows the central tendency, dispersion, and skewness of the data, and it's particularly powerful for identifying outliers.

```python
# My First Boxplot
plt.figure(figsize=(12,6))

plt.boxplot([df['Canada'], df['Japan'], df['USA']],
           tick_labels = ['Cananda', 'Japan', 'USA'])

plt.title('Boxplot of Rates of Selected Countries')
plt.ylabel('Rates')

plt.show()
```

```python
# Improved Boxplot
plt.figure(figsize=(12,6))

plt.boxplot([df['Canada'], df['Japan'], df['USA']],
           tick_labels = ['Canada', 'Japan', 'USA'],
           patch_artist=True, # It allows the boxes in the box plot to be filled with color rather than just outlines.
           boxprops=dict(facecolor='lightblue', color='Blue'), # facecolor specifies that the interior of the bozes will be light blue color.
           medianprops=dict(color='red'),
            whiskerprops=dict(color='red'),
            capprops=dict(color='green'),           
           )

plt.title('Boxplots of Rates for Selected Countries')
plt.ylabel('Rates')
plt.grid(True)
plt.show()
```
