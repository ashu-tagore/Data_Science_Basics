# .loc[]
### Basic Syntax
df.loc[row_selection, column_selection]

### 1. Selecting Rows by Label
#### A. Single row by index label:
- df.loc['row_label']
#### B. Multiple rows by labels
- df.loc[['row1', 'row3', 'row5']]
#### C. Range of rows (inclusive!)
- df.loc['row2':'row5']    ( Includes row2, row3, row4, row5 )

### 2. Selecting Columns by Label
#### A. Single column by index label:
- df.loc['column_label']
#### B. Multiple column by labels
- df.loc[['column1', 'column3', 'column']]
#### C. Range of columns
- df.loc[:, 'colA':'colC']  ( Includes colA, colB, colC )

### 3. Selecting Both Rows and Columns
#### A. Specific Cell
- df.loc['row_label', 'column_name']
#### B.  Multiple rows and columns
- df.loc[['row1', 'row3'], ['colA', 'colC']]
#### C. Range of rows and specific columns
- df.loc['row2':'row4', ['colX', 'colY']]

# Mean (The Average)
- The sum of all values in a dataset divided by the number of values.
- Calculation: Mean = (Sum of all values) / (Number of values)
- Pros:
  - It uses every data point in its calculation, so it represents the entire dataset.
  - It's simple to understand and calculate.
- Cons:
  - It is highly sensitive to outliers (extreme values that are much higher or lower than the rest). A single outlier can drastically skew the mean and make it unrepresentative of the "typical" value.
 
# Median
- The value that separates the higher half of the data from the lower half. It is the literal middle point when all values are sorted in order.
- Calculation:
  - Arrange all values in ascending order.
  - If there is an odd number of values, the median is the middle one.
  - If there is an even number of values, the median is the average of the two middle values.
- Pros:
  - It is robust to outliers. Extreme values do not affect the median.
  - It often gives a better idea of the "typical" value in a skewed dataset.
- Cons:
  - It does not incorporate every data point. It only cares about the value in the middle of the sorted list.

## In Simple Terms
Imagine you have a set of numbers representing something, like the salaries of a group of people.
- Mean (the Average): You add up all the salaries and then divide by the number of people.
- Median (the Middle): You line up all the people from poorest to richest. The person exactly in the middle of that line has the median salary.

```python
df.loc[filt_idiot, 'ConvertedComp'].mean()
# Finds the mean salary of all the people who are from Nepal and have no confidence in their managers
```

## Pandas : Grouping and Aggregation
#### Core Functions: 
- sum(), mean(), count(), min(), max() – Basic summary statistics.

#### .agg() [Aggregate]
- Applies one or more aggregation functions to data.
-  Multiple stats on one column : df['Salary'].agg(['mean', 'sum', 'max', 'min'])
-  Multiple stats on multiple columns : df[['Salary', 'Age']].agg(['mean', 'sum'])
####  .groupby('ColumnName') 
-  Splits the DataFrame into separate groups based on the unique values in 'ColumnName'.
#### The .groupby() + .agg() Combo
- groupby('ColumnName'): Splits the DataFrame into separate groups based on the unique values in 'ColumnName'.
- .agg(): Applies aggregation function(s) to each of those groups separately.

```python
country_group = df.groupby('Country')
country_group
# first this groups your DataFrame by the column Country
# secondly, now your each country e.g. Nepal, India, China becomes its own group
# lastly, country_group is a GroupBy object, not a DataFrame yet
```

```python
# Counting social media users within Groups
country_group['SocialMedia'].value_counts()
```

```python
#  calculates the most popular social media platforms within each country group,
# and then shows the top 50 most frequent individual country-platform combinations.
country_group['SocialMedia'].value_counts().head(50)
```

```python
# filtering for a specific country
country_group['SocialMedia'].value_counts().loc['Nepal']
```

```python
# Alternative way to filter for a specific country
df[df['Country'] == 'Nepal']['SocialMedia'].value_counts()
```

```python
# Plotting bar graph after filtering for a specific country
country_group['SocialMedia'].value_counts().loc['Nepal'].plot.bar()
```

```python
# It calculates the mean (average) annual USD salary for survey respondents from Nepal.
country_group['ConvertedComp'].mean().loc['Nepal']
```

```python
# Calculates the median annual salary for survey respondents from Nepal
country_group['ConvertedComp'].median().loc['Nepal']
```

```python
# Calculates the sum of salaries of all respondents from Nepal
country_group['ConvertedComp'].sum().loc['Nepal']
```

```python
# It calculates the 'mean', 'median', 'min' and 'max' salaries of respondents from all countries
country_group['ConvertedComp'].agg(['mean', 'median', 'min', 'max'])
```

```python
# Calculates the 'mean', 'median', 'min', 'max' of salaries of respondents from Nepal
country_group['ConvertedComp'].agg(['mean','median','min','max']).loc['Nepal']
```

### What is the average number of years developers have been coding professionally? Hint: YearsCodePro
